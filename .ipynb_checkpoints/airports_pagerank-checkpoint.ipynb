```python
%matplotlib inline

import pyspark.sql.functions as F
import pyspark.sql.types as T
import graphframes as GF
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
```

```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
spark = SparkSession.builder.master("local").getOrCreate()
sc = SparkContext.getOrCreate()
```

```python
on_time_df = spark.read.csv('gs://dp-income/2017.csv',header=True)
```

```python
airport_codes_df = spark.read.csv('gs://dp-income/airports.csv',header=True)
airline_ids_df = spark.read.csv('gs://dp-income/airlines.csv',header=True)
```

```python
split_col = F.split(on_time_df['FL_DATE'], '-')
on_time_df = on_time_df.withColumn('Year', split_col.getItem(0))
on_time_df = on_time_df.withColumn('Month', split_col.getItem(1))
on_time_df = on_time_df.withColumn('Day', split_col.getItem(1))
on_time_df = on_time_df.withColumn('YearMonth', F.concat(F.col('Year'),F.lit('-'), F.col('Month')))
```

```python
airport_vertices = (
    airport_codes_df
    .withColumnRenamed('airport_id','id')
    .withColumnRenamed('airport_name','name')
    .select('id','name','city')
)
```

```python
airport_edges = (
    on_time_df
    .select(
        F.col('ORIGIN').alias('src'),
        F.col('DEST').alias('dst'),
        F.col('OP_CARRIER').alias('airline_id'),
        'YearMonth'
    )
    .join(
        airline_ids_df.select('airline_id','airline_name'),
        on='airline_id',
        how='inner'
    )
    .drop('airline_id')
).cache()
```

```python
graph = GF.GraphFrame(airport_vertices, airport_edges)
```

```python
overall_ranks = graph.pageRank(resetProbability=0.01, maxIter=20)

(
    overall_ranks.vertices
    .orderBy("pagerank", ascending=False)
).limit(100).toPandas()
```

```python
(
    overall_ranks.vertices
    .orderBy("pagerank", ascending=False)
).toPandas().head(20)
```

```python
# calculate the monthly "pageranks" of the airports. Normalize all scores so that
# top ranks airport has a score of 1.0, and all other scores are proportional to that.

month_set = [r.YearMonth for r in airport_edges.select('YearMonth').distinct().collect()]

monthly_ranks_list = [
    (
        graph
        .filterEdges('YearMonth = "{0}"'.format(m))
        .pageRank(resetProbability=0.01, maxIter=5)
        .vertices
        .withColumn('YearMonth',F.lit(m).cast(T.StringType()))
    ) for m in month_set
]

monthly_ranks = None

for cur_month_ranks in monthly_ranks_list:
    maxRank = cur_month_ranks.agg(F.max('pagerank').alias('maxRank')).collect()[0].maxRank
    normalized_ranks = cur_month_ranks.withColumn('normalized_rank', F.col('pagerank')/maxRank)
    if monthly_ranks is None:
        monthly_ranks = normalized_ranks
    else:
        monthly_ranks = monthly_ranks.union(normalized_ranks)
monthly_ranks = monthly_ranks.cache()
```

```python
overall_ranks.vertices.repartition(3).write.parquet('gs://dp-income/overall_airport_pageranks', mode='overwrite')
monthly_ranks.repartition(12).write.parquet('gs://dp-income/monthly_airport_pageranks', mode='overwrite')
```

```python
# load saved calculated airport pagerank data and analyze for top N airports

top_airport_count = 10

top_airports = (
    spark.read.parquet('gs://dp-income/overall_airport_pageranks')
    .orderBy("pagerank", ascending=False)
    .limit(top_airport_count)
    .coalesce(1)
)

monthly_ranks_top = (
    spark.read.parquet('gs://dp-income/monthly_airport_pageranks')
    .join(
        F.broadcast(top_airports.select('id')),
        on='id',
        how='inner'
    )
)

airport_set = [(r.name, r.city) for r in top_airports.select('name','city').collect()]

# graph the monthly airport normalized pageranks

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

monthly_ranks_top_pd = monthly_ranks_top.orderBy('YearMonth').toPandas()

fig, ax = plt.subplots()

for airport in airport_set:
    rank_data = monthly_ranks_top_pd[monthly_ranks_top_pd['city'] == airport[1]]
    dates_raw = rank_data['YearMonth']
    dates = [dt.datetime.strptime(date,'%Y-%m') for date in dates_raw]
    ax.plot(
        dates,
        rank_data['normalized_rank'],
        label = '{0} ({1})'.format(airport[0], airport[1]),
        linewidth = 2
    )
    
fig.set_size_inches(16,12)
plt.xlabel("Month")
plt.ylabel("Rank")
plt.setp(plt.legend(loc='lower left', prop={'size': 14}, bbox_to_anchor=(0,1.02,1,0.2)).get_lines(), linewidth=6)    
plt.show()
```
